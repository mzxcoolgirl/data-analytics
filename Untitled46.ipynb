```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
print(df.head())
```

```python
from statsmodels.stats.weightstats import ztest

z_stat, p_value = ztest(df['sepal length (cm)'], value=5.8)
print("Z-stat:", z_stat)
print("P-value:", p_value)
```

```python
from scipy.stats import ttest_ind

group1 = df[df['species'] == 0]['sepal length (cm)']
group2 = df[df['species'] == 1]['sepal length (cm)']

t_stat, p_value = ttest_ind(group1, group2)
print("T-stat:", t_stat)
print("P-value:", p_value)
```

```python
import numpy as np
from scipy.stats import chi2_contingency

df['sepal_cat'] = pd.cut(df['sepal length (cm)'], bins=3, labels=["Short","Medium","Long"])

contingency = pd.crosstab(df['sepal_cat'], df['species'])

chi2, p, dof, expected = chi2_contingency(contingency)

print("Chi2:", chi2)
print("P-value:", p)
```
