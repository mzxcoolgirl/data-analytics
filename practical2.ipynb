```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load an image (replace 'input_image.jpg' with your image file)
# The Image.open() function from Pillow loads the image
# np.array() converts the Pillow image object to a NumPy array
img = np.array(Image.open('bird-8788491_1280.jpg'))

# Display the original image
plt.imshow(img)
plt.title("Original Image")
plt.axis('off') # Hide axes for a cleaner look
plt.show()

print(f"Image shape: {img.shape}")
print(f"Data type: {img.dtype}")
```

greyscaling

```python
# Grayscale conversion using a weighted average formula
# Weights are based on human perception (R: 0.2989, G: 0.5870, B: 0.1140)
grayscale_img = np.dot(img[...,:3], [0.2989, 0.5870, 0.1140])
grayscale_img = grayscale_img.astype(np.uint8) # Convert to 8-bit unsigned integer

plt.imshow(grayscale_img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

Cropped image

```python
# Get image dimensions
height, width, _ = img.shape

# Define crop area (e.g., center half of the image)
start_row, end_row = height // 4, 3 * height // 4
start_col, end_col = width // 4, 3 * width // 4

# Crop the image using slicing
cropped_img = img[start_row:end_row, start_col:end_col]

plt.imshow(cropped_img)
plt.title("Cropped Image")
plt.axis('off')
plt.show()
```

Flipping image

```python
# Flip image horizontally (left-right)
flipped_horizontal = np.fliplr(img)

# Flip image vertically (up-down)
flipped_vertical = np.flipud(img)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(flipped_horizontal)
plt.title("Flipped Horizontally")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(flipped_vertical)
plt.title("Flipped Vertically")
plt.axis('off')
plt.show()
```

adjusting brightness (negative)

```python
# Negative of an RGB image
negative_img = 255 - img

plt.imshow(negative_img)
plt.title("Negative Image")
plt.axis('off')
plt.show()
```

```python
import numpy as np
import imageio.v3 as iio
import matplotlib.pyplot as plt
from skimage.color import rgb2gray, rgba2rgb

if img.shape[-1] == 4:
    img = rgba2rgb(img)

gray = rgb2gray(img).astype(float)
noise_img = gray + 0.9 * gray.std() * np.random.random(gray.shape)

plt.imshow(noise_img, cmap='gray')
plt.axis('off')
plt.title("Noisy Image")
plt.show()
```
